```python
import numpy as np
import pandas as pd 
from numpy import random
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from keras.models import Sequential

```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

<h1>Iris Dataset</h1>

```python
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from tensorflow.keras.utils import to_categorical
train_labels = to_categorical(y_train)
test_labels = to_categorical(y_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.optimizers import Adam
```

<h2> Model "network"</h2>
<p><b>2 layers</b></p>
<ul>
    <li>layer 1: input layer , 512 neurones , activation function relu</li>
    <li>layer 2: output layer , 3 neurones , activation fucntion softmax</li>
</ul>
<p><b>loss: categorical_crossentropy</b></p>

```python
network = Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(4,)))
network.add(layers.Dense(3, activation='softmax'))
```

```python
network.compile(optimizer="adam",
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

```python
network.fit(X_train, train_labels, epochs=20, batch_size=40)
```

```python
test_loss, test_acc = network.evaluate(X_test, test_labels)
#
# Print the test accuracy
#
print('Test Accuracy: ', test_acc, '\nTest Loss: ', test_loss)
```

```python
y_pred=network.predict(X_test)
```

```python
y_pred = np.argmax(y_pred,axis=1)
y_pred
```

<h3>Model "network" metrics</h3>

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```

<h1>Digits Dataset</h1>

```python
import tensorflow as tf
(trainx, trainy), (testx, testy) = tf.keras.datasets.mnist.load_data()
```

```python
trainx = trainx.reshape(trainx.shape[0], 28, 28, 1)
testx = testx.reshape(testx.shape[0], 28, 28, 1)
input_shape = (28, 28, 1)
```

```python
trainx = trainx.astype('float32')
testx = testx.astype('float32')
```

```python
trainx /= 255
testx /= 255
```

<h2> Model "sequential_model"</h2>
<p><b>6 layers</b></p>
<ul>
    <li>layer 1: input layer , 28 neurones , activation function relu</li>
    <li>layer 2: input layer , 28 neurones</li>
    <li>layer 3: input layer , 28 neurones</li>
    <li>layer 4: input layer , 128 neurones , activation function relu</li>
    <li>layer 5: input layer , 128 neurones </li>
    <li>layer 6: output layer , 10 neurones , activation fucntion softmax </li>
</ul>
<p><b>loss:sparse_categorical_crossentropy</b></p>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D
model = Sequential()
model.add(Conv2D(28, kernel_size=(3,3), input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation=tf.nn.relu))
model.add(Dropout(0.2))
model.add(Dense(10,activation=tf.nn.softmax))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x=trainx,y=trainy, epochs=10)
```

<h3>Model "sequential_model" metrics</h3>

```python
model.evaluate(testx, testy)
```

```python
image_index = 2853
plt.imshow(testx[image_index].reshape(28, 28),cmap='Greys')
predict = testx[image_index].reshape(28,28)
pred = model.predict(testx[image_index].reshape(1, 28, 28, 1))
print(pred.argmax())
```

<h2> Model "sequential_model1"</h2>
<p><b>8 layers</b></p>
<ul>
    <li>layer 1: input layer , 64 neurones</li>
    <li>layer 2: input layer , 64 neurones</li>
    <li>layer 3: input layer , 64 neurones</li>
    <li>layer 4: input layer , 128 neurones , activation function relu</li>
    <li>layer 5: input layer , 128 neurones , activation function relu</li>
    <li>layer 6: input layer , 128 neurones , activation function relu</li>
    <li>layer 7: input layer , 128 neurones</li>
    <li>layer 8: input layer , 10 neurones , activation function softmax</li>
</ul>
<p><b>loss:sparse_categorical_crossentropy</b></p>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D
model1 = Sequential()
model1.add(Conv2D(64, kernel_size=(3,3), input_shape=input_shape))
model1.add(MaxPooling2D(pool_size=(2, 2)))
model1.add(Flatten())
model1.add(Dense(128, activation=tf.nn.relu))
model1.add(Dense(128, activation=tf.nn.relu))
model1.add(Dense(128, activation=tf.nn.relu))
model1.add(Dropout(0.2))
model1.add(Dense(10,activation=tf.nn.softmax))
```

```python
model1.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model1.fit(x=trainx,y=trainy, epochs=10)
```

<h3>Model "sequential_model" metrics</h3>

```python
from sklearn.datasets import fetch_olivetti_faces
olivetti = fetch_olivetti_faces()
```

```python
print(np.max(olivetti.data))
print(np.min(olivetti.data))
print(np.mean(olivetti.data))
```

```python
def print_faces(images, target, top_n):
    # set up the figure size in inches
    fig = plt.figure(figsize=(12, 12))
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
    for i in range(top_n):
        # plot the images in a matrix of 20x20
        p = fig.add_subplot(20, 20, i + 1, xticks=[], yticks=[])
        p.imshow(images[i], cmap=plt.cm.bone)
        
        # label the image with the target value
        p.text(0, 14, str(target[i]))
        p.text(0, 60, str(i))
```

```python
print_faces(olivetti.images, olivetti.target, 20)
```

```python
print_faces(olivetti.images, olivetti.target, 400)
```
